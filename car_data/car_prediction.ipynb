```python
import numpy as np
import pandas as pd 
from sklearn.impute import SimpleImputer 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('quikr_car.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df.shape
```

```python
df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
df['name']
```

```python
df.head()
```

```python
df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
df['kms_driven']
```

```python
df = df[df['kms_driven'].str.isnumeric()==True]
```

```python
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df['kms_driven'].isnull().sum()
```

```python
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
df['Price'] = df['Price'].str.replace(',','')
```

```python
price = df.Price[df['Price'].str.isnumeric()==True].astype(int)
price
```

```python
df = df.replace(df.Price[df['Price'].str.isnumeric()==False].values,round(price.median()))
df['Price'] = df['Price'].astype(int)
```

```python
df.isnull().sum()
```

```python
df = df[df['Price']<6e6]
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
pd.set_option('display.max_columns',None) 
pd.set_option('display.max_rows',None) 
```

```python
df['name'].value_counts().count()
```

```python
df['company'].unique()
```

```python
df['fuel_type'] = df['fuel_type'].fillna(df['fuel_type'].mode()[0])
df['fuel_type']
```

```python
df.isnull().sum()
```

```python
df
```

```python
df = df.drop(columns=['company'])
```

```python
x = df[['name','year','kms_driven','fuel_type']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer 
from sklearn.pipeline import make_pipeline 
from sklearn.metrics import r2_score 
```

```python
ohe=OneHotEncoder() 
ohe.fit(x[['name','fuel_type']]) 
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','fuel_type']),remainder='passthrough') 
```

```python
lr=LinearRegression() 
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train) 
```

```python
pipe.score(X_train,y_train)
```

```python
predict = pipe.predict(X_test)
```
